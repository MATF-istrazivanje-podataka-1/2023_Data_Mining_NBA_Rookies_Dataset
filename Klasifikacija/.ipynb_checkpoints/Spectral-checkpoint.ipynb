```python
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('nba_logreg.csv')
```

```python
data=data.fillna(0.0)
```

```python
data=data.drop_duplicates()
```

```python
imenjaci=data.duplicated(subset=['name','gp','min','pts'],keep=False)
```

```python
imenjaci1=data.loc[imenjaci]
imenjaci1.index
```

```python
data.drop(index=(imenjaci1.index),inplace=True)
```

```python
scaler=MinMaxScaler()
```

```python
nazivi=data.columns[1:]
X=data[nazivi]
```

```python
X
```

```python
X = pd.DataFrame(scaler.fit_transform(X), columns=nazivi)
```

```python
X
```

```python
from sklearn.decomposition import PCA
X1=PCA(n_components=2).fit_transform(X)
```

```python
pca_podaci=pd.DataFrame(X1)
pca_podaci
```

```python
feature_names=pca_podaci.columns[0:]
```

```python
from sklearn.metrics import silhouette_score
ks = range(2,8) # [)
inits = ['nearest_neighbors', 'rbf']
fig = plt.figure(figsize=(10,30))
idx = 1
silhouette = []
inertias = []
for k in ks:
    for init in inits:
        spectral = SpectralClustering(n_clusters=k,eigen_solver='arpack',affinity=init)
        spectral.fit(X1)
#         if init == 'nearest_neighbors':
#             inertias.append(spectral.inertia_)
#             silhouette.append(silhouette_score(X1, spectral.labels_))

        fig.add_subplot(len(ks), len(inits), idx)
        idx += 1
        for label in range(k):
            cluster = X1[spectral.labels_ == label]
            plt.scatter(cluster[:,0], cluster[:,1])
        
#         centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_),columns==feature_names)
#         plt.scatter(centroids[0], centroids[1], color='black', marker='x')
            
        plt.title(f'k={k}, init={init}')
        
plt.tight_layout()
```
