```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('nba_logreg.csv')
```

```python
data
```

```python
data=data.fillna(0.0)
```

```python
data=data.drop_duplicates()
```

```python
data
```

```python
imenjaci=data.duplicated(subset=['name','gp','min','pts'],keep=False)
```

```python
imenjaci1=data.loc[imenjaci]
imenjaci1.index
```

```python
data.drop(imenjaci1.index,inplace=True)
data
```

```python
data=data.drop(['3p','fg','reb','ft','name'],axis=1)
```

```python
from autogluon.tabular import TabularPredictor
```

```python
data1=pd.read_csv('nba_logreg.csv')
```

```python
data1
```

```python
data1=data1.drop(['3p','fg','reb','ft','name'],axis=1)
```

```python
train,test=train_test_split(data)
```

```python
model=TabularPredictor(label='target_5yrs').fit(train)
```

```python
model.evaluate(test,silent=True)
```

```python
model.leaderboard(test,silent=True)
```

```python
model1=TabularPredictor(label='target_5yrs').fit(train1)
```

```python
model1.evaluate(test1,silent=True)
```

```python
model1.leaderboard(test1,silent=True)
```
