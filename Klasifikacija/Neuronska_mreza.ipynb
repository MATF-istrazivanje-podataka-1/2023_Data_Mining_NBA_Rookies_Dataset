```python
from sklearn.model_selection import train_test_split
from sklearn import neural_network as nm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_auc_score
```

```python
data=pd.read_csv('nba_logreg.csv')
```

```python
data
```

```python
Y=data['target_5yrs']
X=data.drop('target_5yrs',axis=1)
X=X.drop(['3p','fg','reb','ft','name'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,stratify=Y)
```

```python
model=nm.MLPClassifier(hidden_layer_sizes=2000,activation='logistic',early_stopping=True)
```

```python
model=model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'hidden_layer_sizes': [50,100,200],
          'activation': ['logistic','tanh','relu'],
          'learning_rate':['invscaling','adaptive'],
          'learning_rate_init':[0.001,0.01,0.1],
          'max_iter':[400,800],
          'tol':[0.0004,0.004,0.04,0.4],
          'early_stopping':[True,False]
         }

estimator = GridSearchCV(nm.MLPClassifier(), param_grid=params, cv=2,verbose=2)
```

```python
estimator.fit(x_train,y_train)
```

```python
estimator.best_params_
```
